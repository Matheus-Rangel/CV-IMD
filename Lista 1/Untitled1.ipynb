```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def noisyless(imgs):
    img_out = imgs[0] * 0
    for img in imgs:
        img_out = img_out + img
    img_out = np.floor_divide(img_out, len(imgs))
    return img_out
```

```python
imgs = []
```

```python
for img in os.listdir('Q6/img'):
    imgs.append(cv2.imread('Q6/img/' + img, 0))
```

```python
img = cv2.fastNlMeansDenoisingMulti(imgs, 8, 1)
```
