```python
import cv2
```

```python
import numpy as numpy
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_rgb = cv2.imread('messi.jpg', 1)
img_original = cv2.imread('messi.jpg', -1)
img_grayscale = cv2.imread('messi.jpg', 0)
img = mpimg.imread('messi.jpg')
```

```python
type(img_rgb)
```

```python
%matplotlib inline
```

```python
plt.imshow(img, interpolation='bicubic')
```

```python
img_r = numpy.array(img)
img_g = numpy.array(img)
img_b = numpy.array(img)
for line_pos, line in enumerate(img):
    for pixel_pos, pixel in enumerate(line):
        img_r[line_pos][pixel_pos] = [pixel[0],0,0]
        img_g[line_pos][pixel_pos] = [0,pixel[1],0]
        img_b[line_pos][pixel_pos] = [0,0,pixel[2]]
```

```python
plt.imshow(img_r, interpolation='bicubic')
```

```python
plt.imshow(img_g, interpolation='bicubic')
```

```python
plt.imshow(img_b, interpolation='bicubic')
```

```python
img_result = img_r + img_g + img_b
```

```python
plt.imshow(img_result, interpolation='bicubic')
```
